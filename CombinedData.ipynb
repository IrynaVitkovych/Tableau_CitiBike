```python
# Import setup
import pandas as pd
```

```python
# Create a list containing all the files for 2022 Jersey City data
files = ['Resources/JC-202301-citibike-tripdata.csv',
         'Resources/JC-202302-citibike-tripdata.csv',
         'Resources/JC-202303-citibike-tripdata.csv',
         'Resources/JC-202401-citibike-tripdata.csv',
         'Resources/JC-202402-citibike-tripdata.csv',
         'Resources/JC-202403-citibike-tripdata.csv']
```

```python
# Create an empty dataframe to store all data
citibike_df = pd.DataFrame()
```

```python
# Loop through the list of files and concatonate all data to the dataframe
for file in files:
    data = pd.read_csv(file) 
    citibike_df = pd.concat([citibike_df, data], axis=0)
```

```python
# Show the dataframe
citibike_df
```

```python
# Store the dataframe as CSV file
citibike_df.to_csv('Resources/2023-2024_data.csv', index=False)
```
